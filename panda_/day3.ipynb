```python
d1 = {'key1':[1,2,3,4,5,6],
    'key2':[7,8,9,10,11,12],
    'key3':[13,14,15,16,17,18],
    'key4':[19,20,21,22,23,24]
}
```

```python
## converting dictionary in to dataframe... table format...
## ValueError: All arrays must be of the same length - if one of the list of dict has less elements compare with other lists

import pandas as pd

pd.DataFrame(d1) # return data frame... table with default indexes that it 0,1,2,3,4,5,6

pd.DataFrame(d1, index=['a','b','c','d','e','f']) # return indexes with a,b,c,d,e,f
```

```python
d2 = {'key1':[1,2,3,4,5,"abc"],
    'key2':[7,8,9,"hello",11,12],
    'key3':[13,14,15,16,17,18],
    'key4':[19,20,21,"india",23,"this is"]
}
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)  # by deafult inner merge operations...
## pd.merge(df1, df2, how='inner') - same as above
```

```python
pd.merge(df1,df2, on='key1') 
## this will merge all records from both df which have same 
## records on key1/column  and merge all records from both dfs
```

```python
pd.merge(df1,df2, on='key4') #inner join by default
## this will merge all records from both df which have same 
## records on key4/column  and merge all records from both dfs
## analysis the output key4 difference in both table will not listed.. 
```

```python
pd.merge(df1,df2,how = 'left', on = 'key4') # left join
## this will merge all records from both df which have same 
## records on key14/column  and merge all records from left table with empty on right if key will not match
```

```python
pd.merge(df1,df2,how = 'right', on = 'key4') # right join
## this will merge all records from both df which have same 
## records on key14/column  and merge all records from right table with empty on left if key will not match
```
